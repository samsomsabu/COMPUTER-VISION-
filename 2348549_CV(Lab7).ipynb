```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

**Load the image**

```python
original_image = cv2.imread('/content/Gold Fish.jpg')
```

```python
cv2_imshow(original_image)
```

```python
def add_noise(image, mean=0, std=25):
    row, col, ch = image.shape
    gauss = np.random.normal(mean, std, (row, col, ch))
    gauss = gauss.reshape(row, col, ch)
    noisy_image = image + gauss
    return noisy_image.astype(np.uint8)

noisy_image = add_noise(original_image)
```

```python
cv2_imshow(noisy_image)
```

```python
# Apply Gaussian blur filter to remove Gaussian noise
denoised_gaussian = cv2.GaussianBlur(noisy_image, (5, 5), 0)
```

```python
cv2_imshow( denoised_gaussian)
```

```python
# Apply Median filter to remove salt-and-pepper noise
denoised_median = cv2.medianBlur(noisy_image, 5)
```

```python
cv2_imshow( denoised_median)
```
