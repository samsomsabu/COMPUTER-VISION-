```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Specify the correct path to your image file
image_path = '/content/download.jpg'

# Try reading the image and check if it's loaded successfully
f = cv2.imread(image_path, 0)

if f is None:
    print(f"Error: Unable to read the image from '{image_path}'. Check the file path.")
else:
    # Get the dimensions of the image
    M, N = f.shape

    # Display the image
    plt.imshow(f, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
F = np.fft.fft2(f)
plt.imshow(np.log1p(np.abs(F)), cmap='gray')
plt.axis('off')
plt.show()
```

```python
Fshift = np.fft.fftshift(F)
plt.imshow(np.log1p(np.abs(Fshift)), cmap='gray')
plt.axis('off')
plt.show()
```

```python
H = np.zeros((M, N), dtype=np.float32)
D0 = 50
for u in range(M):
    for v in range(N):
        D = np.sqrt((u - M/2)**2 + (v - N/2)**2)
        if D <= D0:
            H[u, v] = 1
        else:
            H[u, v] = 0
plt.imshow(H, cmap='gray')
plt.axis('off')
plt.show()
```

```python
Gshift = Fshift * H
plt.imshow(np.log1p(np.abs(Gshift)), cmap='gray')
plt.axis('off')
plt.show()
```

```python
G = np.fft.ifftshift(Gshift)
g = np.abs(np.fft.ifft2(G))
plt.imshow(g, cmap='gray')
plt.axis('off')
plt.show()
```

```python
H = 1 - H
plt.imshow(H, cmap='gray')
plt.axis('off')
plt.show()
```

```python
Gshift = Fshift * H
plt.imshow(np.log1p(np.abs(Gshift)), cmap='gray')
plt.axis('off')
plt.show()
```

```python
G = np.fft.ifftshift(Gshift)
g = np.abs(np.fft.ifft2(G))
plt.imshow(g, cmap='gray')
plt.axis('off')
plt.show()
```
