```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# Load the image
image = cv2.imread('/content/Gold Fish.jpg', cv2.IMREAD_GRAYSCALE)
image_float32 = np.float32(image)
```

```python
cv2_imshow(image)
```

```python
# First Order Derivative Methods
# Sobel Operator
sobel_x = cv2.Sobel(image_float32, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image_float32, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.sqrt(cv2.addWeighted(cv2.pow(sobel_x, 2.0), 1.0, cv2.pow(sobel_y, 2.0), 1.0, 0))
```

```python
cv2_imshow(sobel_combined.astype(np.uint8))
```

```python
# Prewitt Operator
prewitt_kernel_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewitt_kernel_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
prewitt_x = cv2.filter2D(image_float32, -1, prewitt_kernel_x)
prewitt_y = cv2.filter2D(image_float32, -1, prewitt_kernel_y)
prewitt_combined = cv2.sqrt(cv2.addWeighted(cv2.pow(prewitt_x, 2.0), 1.0, cv2.pow(prewitt_y, 2.0), 1.0, 0))
```

```python
cv2_imshow( prewitt_combined.astype(np.uint8))
```

```python
# Second Order Derivative Methods
image_uint8 = np.uint8(image)
```

```python
# Laplacian Operator
laplacian = cv2.Laplacian(image_uint8, cv2.CV_8U)
```

```python
cv2_imshow(laplacian)
```

```python
# Optimum Edge Detection Method
canny = cv2.Canny(image, 100, 200)
```

```python
cv2_imshow(canny)
```
