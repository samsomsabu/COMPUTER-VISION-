```python
# Importing cv2
import cv2

# Importing matplotlib.pyplot
import matplotlib.pyplot as plt
```

# **Affine Transformation**

# **Translation**

```python
import numpy as np
```

```python
# Getting the height and width of the image
height, width = img.shape
```

```python
# Defining the shifting coordinates
quart_height, quart_width = height/4, width/4
Coordinates = np.float32([[1, 0, quart_width], [0, 1, quart_height]])
```

```python
# Creating the transformation matrix
img_translation = cv2.warpAffine(img, Coordinates, (width, height))
```

```python
# Plotting the translation image
plt.figure(figsize=(10,10))
plt.subplot(122)
plt.title('Translation')
plt.imshow(img_translation)
plt.show()
```

# **Scaling**

```python
# Setting the grid size
plt.figure(figsize=(20,20))

# Displaying the original image
plt.subplot(131)
plt.title('Original')
plt.imshow(img)
```

```python
# The image is scaled to a smaller size of the original image
resized_image = cv2.resize(img, dsize=(0,0), fx=0.5, fy=0.5)
# Setting the grid size
plt.figure(figsize=(20,20))
# Displaying the resized image
plt.subplot(132)
plt.title('Smaller')
plt.imshow(resized_image)
```

```python
# The image is scaled to a larger size of the original image
resized_image = cv2.resize(img, dsize=(0,0), fx=1.5, fy=1.5)
# Setting the grid size
plt.figure(figsize=(20,20))
# Displaying the resized image
plt.subplot(133)
plt.title('Larger')
plt.imshow(resized_image)
plt.show()
```

# **Average Filter**

```python
from google.colab.patches import cv2_imshow
```

```python
def average_filt(img,kernel_size):
  kernel= np.ones((kernel_size,kernel_size),np.float32)/(kernel_size*kernel_size)
  filt_img=cv2.filter2D(img,-1,kernel)
  return filt_img
```

```python
new_img_avg=average_filt(img,3)
```

```python
from google.colab.patches import cv2_imshow

cv2_imshow(new_img_avg)
```

```python
def laplacian_filt(image):
  kernel= np.array([[0,1,0],[0,-4,1],[0,1,0]],dtype=np.float32)
  filt_img=cv2.filter2D(img,-1,kernel)
  return filt_img
```

```python
new_img=laplacian_filt(img)
```

```python
from google.colab.patches import cv2_imshow

# Display Laplacian filtered image
cv2_imshow(new_img)
```
